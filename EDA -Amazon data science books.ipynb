**1. Which book (title) recieved max reviews?**

```python
max_reviews = dataset['n_reviews'].max() 
review_book = dataset.loc[dataset['n_reviews'] == max_reviews, 'title'].iloc[0]
print(review_book, 'is the most reviewed book with max ', max_reviews, 'reviews')
```

**2. Which book has most percentage of 5 stars?**

```python
five_star = dataset['star5'].max()
five_star_book = dataset.loc[dataset['star5'] == five_star, 'title'].iloc[0]


print(five_star_book, ' got most percentage of 5 stars (', five_star, ')')
```

```python
# Let's find out the number of reviews 
n_reviews_five_star_book = dataset.loc[dataset['title'] == five_star_book, 'n_reviews'].iloc[0]
print(five_star_book, 'got ',n_reviews_five_star_book, ' reviews out of which 5 stars were ', five_star )
```

```python
maximum_reviews = dataset['n_reviews'].max()
five_star_on_max_n_reviews_book = dataset.loc[dataset['title'] == review_book, 'star5'].iloc[0]
print(review_book, 'got', five_star_on_max_n_reviews_book, 'percent of 5 stars and ', maximum_reviews, 'reviews')
```

**3. Which publisher is most famous?**

```python
# Authors vs n_reviews (max and min)
most_reviewed = dataset['n_reviews'].max()
star_author = dataset.loc[dataset['n_reviews'] == most_reviewed, 'author'].iloc[0]

print(star_author, ' got maximum number of reviews (', most_reviewed, ')')
```

```python
least_reviewed = dataset['n_reviews'].min()
_author = dataset.loc[dataset['n_reviews'] == least_reviewed, 'author'].iloc[0]

print(_author, ' got least number of reviews (', least_reviewed, ')')
```

 **5. What are the average number of pages in a data science book and the costliest book?**

```python
average_no_pages = dataset['pages'].mean().round()    # round function is used to roundoff the value.
print('Average pages per book are - ',average_no_pages)
```

```python
costliest = dataset['price'].max()
costliest_book = dataset.loc[dataset['price'] == costliest, 'title'].iloc[0]

print(costliest_book, 'is the costliest book priced at $', costliest)
```

```python
least_cost = dataset['price'].min()
least_costliest_book = dataset.loc[dataset['price'] == least_cost, 'title'].iloc[0]

print(least_costliest_book, 'is the cheapest book priced at $', least_cost)
```
